# 제출 정답

```python
nums = list(range(1,31))

for i in range(len(nums)-2):
    n = int(input())
    if n in nums:
        nums.remove(n)
# 출력이 두줄임을 잊었다.
# 과제 제출한 학생은 28명으로 고정되어있고, 2명의 미제출자가 확정이다.
# nums는 1부터 30까지 순서대로 (range)로 정의 되어 있기때문에
# 제출한 모든 학생을 제거하면 작은 순서대로 nums의 요소가 남기 때문에
# 인덱스로 표시하면 될 것 같다.
print(nums[0])
print(nums[1])
```
